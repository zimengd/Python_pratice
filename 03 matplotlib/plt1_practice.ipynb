```python
import pandas as pd
unrate = pd.read_csv('unrate.csv')
unrate.head() #失业率
```

# pd.to_datetime()

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate['DATE'].head()
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
first_twelve = unrate[0:12]
plt.plot(first_twelve['DATE'],first_twelve['VALUE']) # x,y
plt.show
```

# plt.xticks(rotation = 45)

```python
plt.plot(first_twelve['DATE'],first_twelve['VALUE'])
plt.xticks(rotation = 45) #rotate x-tick by 45 degrees
plt.show()
```

# xlable()
# ylable()
# title()

```python
#xlabel(): accepts a string value, which gets set as the x-axis label.
#ylabel(): accepts a string value, which is set as the y-axis label.
#title(): accepts a string value, which is set as the plot title.
plt.plot(first_twelve['DATE'], first_twelve['VALUE'])
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```
